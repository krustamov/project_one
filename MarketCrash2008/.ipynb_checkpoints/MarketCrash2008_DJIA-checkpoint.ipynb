### Import dependencies

```python
import matplotlib.pyplot as plt
from matplotlib import style
import datetime as dt
import pandas as pd
import os
%matplotlib inline
```

### Set matplotlib style

```python
style.use('fivethirtyeight')
```

### Read csv files into a dataframe

```python
csvpath1 = os.path.join("Stock Prices\dji.csv")
df = pd.read_csv(csvpath1)
```

### Convert dates to datetime

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by='Date')
```

### Create new dataframe with a narrower date range to examine stock market crash of 2008

```python
dow = df[(df['Date'] > '2006-12-31') & (df['Date'] < '2013-3-1')]
```

### Assign column values to lists

```python
date = dow['Date'].tolist()
average = dow['Close'].tolist()
```

### Graph the DJIA response to the crash annotating key events

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)

Average, = plt.plot(date, average, color="blue",linewidth=1, label="Average")

plt.fill_between(date, average, y2=0, where=None, interpolate=False, alpha=0.3, step=None, hold=None, data=None)

ax.annotate('9.15.08: Lehman Brothers Declares Bankruptcy', xy=('2008-09-15', 13000.00),
            arrowprops=dict(facecolor='blue', shrink=0.5),)

ax.annotate('9.29.08: Senate Denies Bank Bailout', xy=('2008-09-29', 12000.00),
            arrowprops=dict(facecolor='red', shrink=0.5),)

ax.annotate('2.17.09: Economic Stimulus Package Passed', xy=('2009-2-17', 8900.00),
            arrowprops=dict(facecolor='green', shrink=0.5),)

ax.annotate('3.17.08: Fed Funds Bear Stearns Buyout', xy=('2008-3-17', 14000.00),
            arrowprops=dict(facecolor='yellow', shrink=0.5),)

ax.annotate('9.7.08: US Govt. Bails Out Fannie Mae & Freddie Mac', xy=('2008-09-07', 6000.00),
            arrowprops=dict(facecolor='orange', shrink=0.5),)

plt.title("DJIA Response to 2008 Market Crash (7-Year-Span)")
plt.xlabel("Date")
plt.ylabel("Closing Average")
plt.savefig("DJIA_7yr_span.png")
```
