### Import dependencies

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
from matplotlib import style
import datetime as dt
import pandas as pd
import os
%matplotlib inline
```

### Set Matplotlib style

```python
style.use('fivethirtyeight')
```

### Read csv files into a dataframe

```python
csvpath1 = os.path.join("Stock Prices","aal.csv")
csvpath2 = os.path.join("Stock Prices","aapl.csv")
csvpath3 = os.path.join("Stock Prices","gs.csv")
csvpath4 = os.path.join("Stock Prices","hd.csv")
csvpath5 = os.path.join("Stock Prices","ibm.csv")
csvpath6 = os.path.join("Stock Prices","jnj.csv")
csvpath7 = os.path.join("Stock Prices","pg.csv")
csvpath8 = os.path.join("Stock Prices","wmt.csv")
csvpath9 = os.path.join("Stock Prices","xom.csv")

aal_1 = pd.read_csv(csvpath1)
appl_1 = pd.read_csv(csvpath2)
gs_1 = pd.read_csv(csvpath3)
hd_1 = pd.read_csv(csvpath4)
ibm_1 = pd.read_csv(csvpath5)
jnj_1 = pd.read_csv(csvpath6)
pg_1 = pd.read_csv(csvpath7)
wmt_1 = pd.read_csv(csvpath8)
xom_1 = pd.read_csv(csvpath9)
```

### Convert dates to datetime

```python
aal_1['Date'] = pd.to_datetime(aal_1['Date'])
aal_1 = aal_1.sort_values(by='Date')

appl_1['Date'] = pd.to_datetime(appl_1['Date'])
appl_1 = appl_1.sort_values(by='Date')

gs_1['Date'] = pd.to_datetime(gs_1['Date'])
gs_1 = gs_1.sort_values(by='Date')

hd_1['Date'] = pd.to_datetime(hd_1['Date'])
hd_1 = hd_1.sort_values(by='Date')

ibm_1['Date'] = pd.to_datetime(ibm_1['Date'])
ibm_1 = ibm_1.sort_values(by='Date')

jnj_1['Date'] = pd.to_datetime(jnj_1['Date'])
jnj_1 = jnj_1.sort_values(by='Date')

pg_1['Date'] = pd.to_datetime(pg_1['Date'])
pg_1 = pg_1.sort_values(by='Date')

wmt_1['Date'] = pd.to_datetime(wmt_1['Date'])
wmt_1 = wmt_1.sort_values(by='Date')

xom_1['Date'] = pd.to_datetime(xom_1['Date'])
xom_1 = xom_1.sort_values(by='Date')
```

### Create new dataframe with a narrower date range to examine stock market crash of 2008

```python
aal =  aal_1 [(aal_1['Date'] > '2006-12-31') & (aal_1['Date'] < '2013-12-31')]
aapl = appl_1[(appl_1['Date'] > '2006-12-31') & (appl_1['Date'] < '2013-12-31')]
gs =   gs_1  [(gs_1['Date'] > '2006-12-31') & (gs_1['Date'] < '2013-12-31')]
hd =   hd_1  [(hd_1['Date'] > '2006-12-31') & (hd_1['Date'] < '2013-12-31')]
ibm =  ibm_1 [(ibm_1['Date'] > '2006-12-31') & (ibm_1['Date'] < '2013-12-31')]
jnj =  jnj_1 [(jnj_1['Date'] > '2006-12-31') & (jnj_1['Date'] < '2013-12-31')]
pg =   pg_1  [(pg_1['Date'] > '2006-12-31') & (pg_1['Date'] < '2013-12-31')]
wmt =  wmt_1 [(wmt_1['Date'] > '2006-12-31') & (wmt_1['Date'] < '2013-12-31')]
xom =  xom_1 [(xom_1['Date'] > '2006-12-31') & (xom_1['Date'] < '2013-12-31')]
```

### Create datafames to zoom in on specific date range

```python
aal2 =  aal_1 [(aal_1['Date'] > '2007-12-31') & (aal_1['Date'] < '2009-12-31')]
aapl2 = appl_1[(appl_1['Date'] > '2007-12-31') & (appl_1['Date'] < '2009-12-31')]
gs2 =   gs_1  [(gs_1['Date'] > '2007-12-31') & (gs_1['Date'] < '2009-12-31')]
hd2 =   hd_1  [(hd_1['Date'] > '2007-12-31') & (hd_1['Date'] < '2009-12-31')]
ibm2 =  ibm_1 [(ibm_1['Date'] > '2007-12-31') & (ibm_1['Date'] < '2009-12-31')]
jnj2 =  jnj_1 [(jnj_1['Date'] > '2007-12-31') & (jnj_1['Date'] < '2009-12-31')]
pg2 =   pg_1  [(pg_1['Date'] > '2007-12-31') & (pg_1['Date'] < '2009-12-31')]
wmt2 =  wmt_1 [(wmt_1['Date'] > '2007-12-31') & (wmt_1['Date'] < '2009-12-31')]
xom2 =  xom_1 [(xom_1['Date'] > '2007-12-31') & (xom_1['Date'] < '2009-12-31')]
```

### Assign column values to lists for first graph

```python
date = aal['Date'].tolist()
aal_price = aal['Close'].tolist()
aapl_price = aapl['Close'].tolist()
gs_price = gs['Close'].tolist()
hd_price = hd['Close'].tolist()
ibm_price = ibm['Close'].tolist()
jnj_price = jnj['Close'].tolist()
pg_price = pg['Close'].tolist()
wmt_price = wmt['Close'].tolist()
xom_price = xom['Close'].tolist()
```

### Assign column values to lists for second graph

```python
date2 = aal2['Date'].tolist()
aal2_price = aal2['Close'].tolist()
aapl2_price = aapl2['Close'].tolist()
gs2_price = gs2['Close'].tolist()
hd2_price = hd2['Close'].tolist()
ibm2_price = ibm2['Close'].tolist()
jnj2_price = jnj2['Close'].tolist()
pg2_price = pg2['Close'].tolist()
wmt2_price = wmt2['Close'].tolist()
xom2_price = xom2['Close'].tolist()
```

### Graphically compare stocks before & after crisis, annotating specific events

```python
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(111)

ax.annotate('9.15.08: Lehman Brothers', xy=('2008-09-15', 180.00),
            arrowprops=dict(facecolor='blue', shrink=0.5),)

ax.annotate('9.29.08: Bailout Vetoed', xy=('2008-09-29', 195.00),
            arrowprops=dict(facecolor='red', shrink=0.5),)

ax.annotate('2.17.09: Stimulus Package', xy=('2009-2-17', 210.00),
            arrowprops=dict(facecolor='green', shrink=0.5),)

ax.annotate('3.17.08: Bear Stearns Buyout', xy=('2008-3-17', 225.00),
            arrowprops=dict(facecolor='yellow', shrink=0.5),)


plt.plot(date, aal_price, color="blue", linewidth=1, label="American Airlines")
plt.plot(date, aapl_price, color="black", linewidth=1, label="Apple")
plt.plot(date, gs_price, color="red", linewidth=1, label="Goldman Sachs")
plt.plot(date, hd_price, color="purple", linewidth=1, label="Home Depot")
plt.plot(date, ibm_price, color="green", linewidth=1, label="IBM")
plt.plot(date, jnj_price, color="yellow", linewidth=1, label="Johnson & Johnson")
plt.plot(date, pg_price, color="orange", linewidth=1, label="Proctor & Gamble")
plt.plot(date, wmt_price, color="cyan", linewidth=1, label="Wal-Mart")
plt.plot(date, xom_price, color="pink", linewidth=1, label="Exxon Mobil")


plt.legend(loc=2, shadow=bool)
plt.title("Stock Market Analysis: 2008 Stock Market Crash - 7 Year Span")
plt.xlabel("Date")
plt.ylabel("(Stock Price)")

#plt.savefig("2008_crash_7yr_span.png")
```
